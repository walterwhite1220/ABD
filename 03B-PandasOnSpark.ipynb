```
# import pandas for spark
# imports
# hides all warnings
import warnings
warnings.filterwarnings('ignore')
# import pandas for spark
import pyspark.pandas as ps
# pandas 
import pandas as pd
# numpy
import numpy as np
```

```
# reade file nifty.csv
# Read CSV

# File location and type
file_location = "/FileStore/tables/test/nifty.csv"
file_type = "csv"

# CSV options
infer_schema = "true"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
# convert dataframe to pandas spark dataframe
psdf = ps.DataFrame(df)
# display row & col count
# rows & cols
print("\n*** Rows & Cols ***")
print("Rows",psdf.shape[0])
print("Cols",psdf.shape[1])
```
